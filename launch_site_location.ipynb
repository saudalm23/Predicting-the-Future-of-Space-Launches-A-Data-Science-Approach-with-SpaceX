Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
# Create base map centered at Cape Canaveral
site_map_clustered = folium.Map(location=[28.5, -80.6], zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map_clustered)

# Loop through each launch record and add to the cluster
for _, row in spacex_df.iterrows():
    coord = [row['Lat'], row['Long']]
    outcome = 'Success' if row['class'] == 1 else 'Failure'
    color = 'green' if row['class'] == 1 else 'red'
    
    popup_text = f"Launch Site: {row['Launch Site']}<br>Outcome: {outcome}"
    
    folium.Marker(
        location=coord,
        popup=popup_text,
        icon=folium.Icon(color=color)
    ).add_to(marker_cluster)
```

```python

from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Your existing haversine function
from math import sin, cos, sqrt, atan2, radians
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Coordinates
launch_site = [28.5623, -80.5774]
city = [28.6100, -80.8070]
railway = [28.5721, -80.5852]
highway = [28.5638, -80.5701]

# Create map
site_map = folium.Map(location=launch_site, zoom_start=11)

# Add launch site marker
folium.Marker(
    launch_site,
    popup="Launch Site",
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Function to add line + marker with distance label
def add_proximity_marker(start, end, label, color):
    dist = calculate_distance(start[0], start[1], end[0], end[1])
    
    # Line
    folium.PolyLine([start, end], color=color, weight=2.5, opacity=1).add_to(site_map)
    
    # Marker at proximity point
    folium.Marker(end, popup=label, icon=folium.Icon(color=color)).add_to(site_map)
    
    # Distance label
    folium.Marker(
        location=end,
        icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:{color};"><b>{label}: {dist:.2f} KM</b></div>'
        )
    ).add_to(site_map)

# Add proximity features
add_proximity_marker(launch_site, city, 'City', 'orange')
add_proximity_marker(launch_site, railway, 'Railway', 'green')
add_proximity_marker(launch_site, highway, 'Highway', 'purple')

# Show map
site_map
```
